# Categorical data

```python
# countplot
```

```python
suv = df['Survived'].value_counts()
print(suv)
suv.plot(kind='bar')
```

```python
p =df['Pclass'].value_counts()
print(p)
p.plot(kind='bar')
```

```python
e = df['Embarked'].value_counts()
print(e)
e.plot(kind='pie',autopct='%.2f')
```

```python
s =df['Sex'].value_counts()
print(s)
s.plot(kind='bar')
```

```python
s =df['Sex'].value_counts()
print(s)
s.plot(kind='pie')
```
